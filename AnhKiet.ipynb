```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./datasets/HR_Analytics.csv")
df.head()
```

## Biểu đồ 1:

**Tiêu đề:** Tỉ trọng giới tính của các nhân viên trong tổ chức

**Loại biểu đồ:** Pie Chart

**Lý do lựa chọn:** Để so sánh về tỉ lệ phần trăm giữa nam và nữ trong tổ chức.

**Trực quan hóa:**

```python
explode = (0, 0.1)
plt.pie(df['Gender'].value_counts(), labels=df['Gender'].value_counts().index,
        explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Proportion of employees by Gender')
plt.show()
```

**Nhận xét:** Các nhân viên nam trong công ty chiếm tỉ trọng nhiều hơn giới tính nữ, và hơn 20%.

## Biểu đồ 2:

**Tiêu đề:** Tình trạng hôn nhân có liên quan gì với quyết định thôi việc hay không?

**Loại biểu đồ:** Countplot

**Lý do lựa chọn:** Để tìm ra mối liên hệ giữa tình trạng hôn nhân và quyết định thôi việc ở nhân viên.

**Trực quan hóa:**

```python
plt.rcParams["figure.figsize"] = [8, 5]
plt.rcParams["figure.autolayout"] = True
g = sns.countplot(x='MaritalStatus', hue='Attrition', data=df)
g.set(title='Number of attritional employees or not by Marital Status')
g.set_xlabel('Marital Status')
g.set_ylabel('Number of employees')
g.set_yticks(range(0, 700, 100))
for p in g.patches:
   g.annotate('{}'.format(p.get_height()), (p.get_x()+0.15, p.get_height()+5))
plt.show()
```

**Nhận xét:**
- Số lượng nhân viên đã kết hôn là nhiều nhất, lần lượt tiếp đến là đã ly hôn và độc thân.
- Trong tất cả tình trạng hôn nhân, số lượng nhân viên không bị hao hụt luôn nhiều hơn số lượng nhân viên bị hao hụt.
- Khoảng cách giữa số lượng nhân viên không bị hao hụt và bị hao hụt lớn nhất ở nhóm đã kết hôn, tiếp đến là đã ly hôn và nhỏ nhất là độc thân.
Từ những nhận xét trên, ta nhận định những nhân viên đã kết hôn và đã ly hôn sẽ gắn bó với công ty hơn so với nhóm độc thân.
