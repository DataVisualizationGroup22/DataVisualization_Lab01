## Biểu đồ 1

**Tiêu đề:** Mối tương quan giữa thời gian làm việc, mức lương, tỉ lệ hài lòng với công việc và tình trạng nghỉ việc của nhân viên.

**Loại biểu đồ:** Bubble plot(scatter plot with varying size) + một số cải tiến.

**Lý do lựa chọn:** Để tìm hiểu được sự tác động của mối quan hệ giữa lương và tổng thời gian làm việc của nhân viên lên sự hài lòng đối với công việc của họ thì việc trực quan bằng Bubble plot là cần thiết. Hai chiều của bubble plot sẽ thể hiện mối quan hệ giữa lương và thời gian làm việc, kích thước của điểm dữ liệu thể hiện mức độ hài lòng của nhân viên và có thể phân loại thông qua hue=Attrition.

**Trực quan hóa:**

Để có thể dùng Attrition làm biến phân loại trong đồ thị thì cần chuyển giá trị của thuộc tính này về dạng số.

```python
from sklearn.preprocessing import LabelEncoder

df_copy=df.copy()
label_encoder=LabelEncoder()
new_col=label_encoder.fit_transform(df_copy['Attrition'].values)
df_copy['Attrition_num']=new_col
```

```python
df_copy.head()
```

```python
sns.relplot(x='TotalWorkingYears', y='MonthlyIncome', hue='Attrition_num', size='JobSatisfaction',
            sizes=(50, 500), alpha=.5, palette="magma", height=5, data=df_copy)
plt.title("The impact of the relationship" + "\n"
          "between slary and working time on job satisfaction", fontweight='bold');
```

**Nhận xét:** Từ biểu đồ ta thấy:
- Mức đồ hài lòng của nhân viên về công việc thường nằm ở mức 3. Cho thấy với những nhân viên lâu năm (lớn hơn 5 năm) hay các nhân viên có thời gian làm việc ít hơn vẫn tương đối hài lòng với công việc, không quá tiêu cực nhưng cũng không đánh giá quá 'màu hồng' về công việc.
- Tuy nhiên sự phân bố của thuộc tính 'sự hài lòng' rất rộng, do đó không thể rút ra những kết luận nào mang tính quá quy luật cho thuộc tính này.
- Số lượng nhân viên rời đi là rất ít hơn so với số lượng ở lại, và trong những nhân viên rời đi cũng có khá nhiều đánh giá không hài lòng về công việc.
- Những mức lương rất cao chỉ xuất hiện ở những nhân viên có tổng thời gian làm việc trên 15 năm, và hầu như không có tình trạng rời công ty ở nhóm nhân viên này.
- Tình trạng nghỉ việc của nhân viên thường xuất hiện nhiều ở mức lương dưới 10.000$, đều này cũng khá thực tế khi dường như có thể dễ cân nhắc 'nhảy việc' giữa những đề nghị ở khoảng lương này từ nhiều nới có điều kiện tốt khác nhau cho công việc.

## Biểu đồ 2

**Tiêu đề:** Mối quan hệ giữa tỉ lệ rời đi và sự hài lòng về môi trường làm việc.

**Loại biểu đồ:** Bar chart + line plot.

**Lý do lựa chọn:** Dùng line plot đếm số lượng nhân viên theo từng nhóm đánh giá hài lòng về môi trường làm việc ở 4 mức 1, 2, 3, 4. Sau đó tính tỉ lệ rời đi trong nhóm đó và vẽ các tỉ lệ bằng bar chart.

**Trực quan hóa:**

```python
new_df=df.groupby('EnvironmentSatisfaction').count()
idx=list(new_df.index) #get groups as group 1, group 2, group 3, group 4 
count=list(new_df['Age'].values)
```

```python
percent_dropout=[] #list contains employees dropout percent
for i in range(len(idx)):
    percent_dropout.append(round((len(df[df['EnvironmentSatisfaction']==idx[i]][df['Attrition']=='Yes'])/count[i])*100,3))
percent_dropout
```

```python
envSat_attrition_df={idx[i]:[count[i],percent_dropout[i]] for i in range(len(idx))}
envSat_attrition_df['Thống kê']=['Số lượng nhân viên','Phần trăm nghỉ việc']
envSat_attrition_df=pd.DataFrame(envSat_attrition_df).set_index('Thống kê')
envSat_attrition_df.head()
```

```python
groups=['1','2','3','4']
fig,ax1=plt.subplots(figsize=(13,5))# bar chart
bars=ax1.bar(groups,envSat_attrition_df.loc[['Phần trăm nghỉ việc']].values[0])
for rect in bars:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2.0, height, f'{height:.3f}', ha='center', va='bottom')
ax2=ax1.twinx()
ax2.plot(groups,envSat_attrition_df.loc[['Số lượng nhân viên']].values[0],lw=1,marker='o',color='r')
ax1.set(xlabel='Nhóm giá trị đánh giá', ylabel='Phần trăm nghỉ việc (%)')
ax2.set(ylabel='Số lượng nhân viên (người)')
ax1.set_ylim(0,30)
ax2.set_ylim(0,700)
ax1.set_title(f'Tương quan giữa tỉ lệ rời đi và sự hài lòng về môi trường làm việc trong các nhóm đánh giá', size=15)
fig.tight_layout()
plt.show();
```

**Nhận xét:** Từ biểu đồ ta thấy:
- Số lượng nhân viên hài lòng về môi trường làm việc cũng rất cao tương tự hài lòng về công việc.
- Tỉ lệ rời đi trung bình ở các nhóm đánh giá thấp về môi trường làm việc cao hơn so với các nhóm đánh giá cao. Điều này cũng rất hợp lí với thực tế.
- Từ đây cho thấy bên cạnh tiền lương thì yếu tố môi trường làm việc cũng là yếu quan trọng quyết định lớn đến việc nhân viên có thật sự gắn bó lâu dài hay không.
